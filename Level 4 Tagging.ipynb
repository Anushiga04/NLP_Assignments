Let's try to understand some **Parts Of Speech**

1. **Possesion Noun (NN\$)** - any word which ends with **'s** 
    > ***Jennifer’s*** new handbag matches her shoes perfectly
2. **Gerund (VBG)** - any word which ends with **ing** 
    > ***Swimming*** in the ocean has been Sharon’s passion since she was five years old.
3. **Simple Past (VBD)** - any word which ends with **ed**
    > We ***crossed*** the Channel yesterday.
4. **Modal (MD)** - any word which ends with **ould**
    > He **should** not be late
5. **Third Singular Present (VBZ)** - any word which ends with **es**     
    > He **misses** her so much
6. **Numbers (CD)** - any number integer or decimal
    > 24.90
7. **Plural Nouns (NNS)** - any word ending with **s**
    > You stole my **ideas** and didn’t give me any credit.

# Tagging the content

```python
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize
from collections import Counter
```

```python
file = open('/home/anushya/Documents/wilson_doc.txt')
text = file.read()
text
```

# PART - 1

```python
tokenizer = word_tokenize(text)
tokenizer
```

```python
#Tagging th words in content
tagging = nltk.pos_tag(tokenizer)
tagging
```

```python
list(zip(*tagging))[0]
```

```python
#Finding the most common tags
tag_fd = nltk.FreqDist(tag for (word, tag) in tagging)
tag_fd.most_common()
```

```python
#Default tagging the content 
default_tagger = nltk.DefaultTagger('NN')
default_tagger.tag(tokenizer)
```

# PART - 2

```python
#Defining own tags
patterns = [
    (r'.*ing$', 'VBG'),               # gerunds
    (r'.*ed$', 'VBD'),                # simple past
    (r'.*es$', 'VBZ'),                # 3rd singular present
    (r'.*ould$', 'MD'),               # modals
    (r'.*\'s$', 'NN$'),               # possessive nouns
    (r'.*s$', 'NNS'),                 # plural nouns
    (r'^-?[0-9]+(.[0-9]+)?$', 'CD'),  # cardinal numbers
    (r'.*', 'NN')]                     # nouns (default)
 
```

```python
sentence_1 = 'Jennifer’s new handbag matches her shoes perfectly'
regexp_tagger = nltk.RegexpTagger(patterns)
regexp_tagger.tag(word_tokenize(sentence_1))
```
